# clean the data

```python
# drop all columns that contain the string Kommentar
clean_data = raw_data.drop(raw_data.filter(regex='Kommentar').columns, axis=1)
```

```python
clean_data.info()
```

```python
clean_data.head()
```

```python
# replace all Ja with 1 and all Nein with 0
replacement_dict = {'Ja' : 1, 'Nein' : 0}
clean_data = clean_data.replace(replacement_dict)
```

```python
clean_data.head()
```

```python
# drop all columns that exclusively contain NaN
print(len(clean_data.columns[clean_data.isna().all()]))
clean_data = clean_data.drop(clean_data.columns[clean_data.isna().all()], axis=1)
```

```python
clean_data.head()
```

```python
clean_data.dtypes
```

## work on the object columns

```python
object_columns = clean_data.select_dtypes(include='object')
```

```python
object_columns.head()
```

```python
object_columns.info()
```

```python
len(object_columns.filter(regex='Sonstiges').columns)
```

```python
object_columns.filter(regex='Sonstiges')
```

```python
object_columns = object_columns.drop(object_columns.filter(regex='Sonstiges'), axis=1)
```

```python
object_columns.head()
```

```python
object_columns = object_columns.drop(object_columns[['submitdate. Datum Abgeschickt', 'startlanguage. Start-Sprache']], axis=1)
```

```python
object_columns.head()
```

```python
object_columns['AP1. Wie alt sind Sie?'].unique()
```

```python
replacement_dict_Alter = {'Unter 25' : 0, 
                          '25–29' : 1, 
                          '30–34' : 2, 
                          '35–39' : 3, 
                          '40–44' : 4, 
                          '45–49' : 5,
                          '50–54' : 6,
                          '55–59' : 7,
                          '60–64' : 8,
                          '65–70' : 9,
                          'Älter als 70' : 10
                          }

object_columns['AP1. Wie alt sind Sie?'] =  object_columns['AP1. Wie alt sind Sie?'].replace(replacement_dict_Alter)
```

```python
object_columns.head()
```

```python
object_columns.nunique()
```

### check out whether there are columns that can be one hot encoded

```python
# Calculate number of unique values for each column
unique_counts = object_columns.nunique()

# Create a new DataFrame with the results
unique_counts_df = pd.DataFrame(unique_counts).reset_index()
unique_counts_df.columns = ['Column Name', 'Number of Unique Values']

unique_counts_df
```

```python
unique_counts_df = unique_counts_df.sort_values(by='Number of Unique Values')
```
